```python
# simple example :  

data_tips = sns.load_dataset('tips').head(10)  # Restaurant tipping data
print(data_tips,"\n\n\n")

sns.pairplot(data = data_tips)
plt.show()

# so we will se 

# 1 :  size versus total_bill
# 1 :  size versus tip
# 1 :  tip versus total_bill
  

# This is what we called pair plotting.


# Bydefault the  pairplot prints all numeriucal value plotting . But if we want plots of some specific plots of numerical groups we can do for sure.
```

```python

# vars parameter:
# Bydefault the  pairplot prints all numeriucal value plotting . But if we want plots of some specific plots of numerical groups we can do for sure.
# To do it just pass   vars = [ "col_a", "col_b"]. only that col name that you wanted to make pair plot.


# simple example :  

data_tips = sns.load_dataset('tips').head(10)  # Restaurant tipping data
print(data_tips,"\n\n\n")

sns.pairplot(data = data_tips, vars = [ "tip", "size"]) # got it.
plt.show()

# here we seen ! 


# 1: tip vss size and size vs tip as 
```

```python
# hue parameter:   used for specific pairing okay!







data_tips = sns.load_dataset('tips').head(10)  # Restaurant tipping data
#print(data_tips,"\n\n\n")

sns.pairplot(data = data_tips,hue = "sex") # got it.
plt.show()

# things gives more leverage to us!
```

```python
# using hue and  vars same time:


data_tips = sns.load_dataset('tips').head(10)  # Restaurant tipping data
#print(data_tips,"\n\n\n")

sns.pairplot(data = data_tips,vars = ["tip","size"],hue = "sex") # got it.
plt.show()

# soeted hai!
```

```python
# hue_order and palette : 

# 1: hue_order = [  ]: its basically decides the order of elemnts of hue wale group or column , ki kaun sa hua object ka name pahele apppear hoga.
# palette : Ye to pata hai!




data_tips = sns.load_dataset('tips').head(10)  # Restaurant tipping data
#print(data_tips,"\n\n\n")

sns.pairplot(data = data_tips,hue = "sex", hue_order = ["Female", "Male"] , palette = "viridis") # got it.
plt.show()


# Got it !

```

```python
# order = [ ] parameter : Its basically deals with the order of written elemnts name on the horix=zontal axis. 
# Initiallly order was Tergersen, Biscoe island, Dream  but as we change it we it becomes as Dream , biscoe island, Torgersen.


data_penguins = sns.load_dataset('penguins')  # Penguin species data
#print(data_penguins)

sns.boxplot(x= "island", y = "bill_length_mm", data = data_penguins,hue = "species", palette = "dark:m",order = ["Dream", "Biscoe island","Torgersen"] )
plt.show()
```

```python
# showmeans( yes showmeans  not   showmean ) parameter : Its basically used to show the mean value of data via any marker see it 
# to enable it just pass showmean = True.as_integer_ratio
#  showmeans = False is bydefault.


data_penguins = sns.load_dataset('penguins')  # Penguin species data
#print(data_penguins)

sns.boxplot(x= "island", y = "bill_length_mm", data = data_penguins,hue = "sex", palette = "dark:r", showmeans = True)  # here green traingular markers shows the  mean value.
plt.show()
```

```python
# meanprops = {"marker = " ", "markeredgecolor": "  "}  ; These parameter used to deal withy the texture of marker of that showmean marker  and its edgecolors.




data_penguins = sns.load_dataset('penguins')  # Penguin species data
#print(data_penguins)

sns.boxplot(x= "island", y = "bill_length_mm", data = data_penguins,hue = "sex", palette = "dark:r", showmeans = True, meanprops = {"marker":"<", "markeredgecolor": "m"})  # here green traingular markers shows the  mean value.
plt.show()
```

```python
# linewidth parameter : Its basically use to deal with the width of line of box 



data_penguins = sns.load_dataset('penguins')  # Penguin species data
#print(data_penguins)

sns.boxplot(x= "island", y = "bill_length_mm", data = data_penguins,hue = "sex", palette = "dark:r", showmeans = True, linewidth = 4)  # box line is more thicker due to high value of the linewidth .
plt.show()
```

```python
# palette parameter!


data_penguins = sns.load_dataset('penguins')  # Penguin species data
#print(data_penguins)

sns.boxplot(x= "island", y = "bill_length_mm", data = data_penguins,hue = "sex", showmeans = True, palette = "plasma")  # here green traingular markers shows the  mean value.
plt.show()
```

```python
# orient parameetr : Its basicalys deals with the whetrer  the graph will be vertical or horizontal! 

# conditions to use the orinent parameter : to use it make sure your both x and y column have numrerical data  coz  non numerical column data is unacceptable here.

# for vertical use orient = "v"

# for horizontal use orient = "h".  here orient = "v" is bydefault.
```
