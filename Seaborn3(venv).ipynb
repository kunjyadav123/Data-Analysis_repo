```python
# vmin   and vmax  parameter : 
# these basically do changes the colorbar ranges with sepcified values vmin and vmax.
# vmin specifes lowest colorbar value.
# vmax specifies highest colorbar value.



iris = sns.load_dataset('iris')  # Flower species data
data_2 = iris.drop(columns = ["species"], axis = 0).head(20)

# print(data_1) # ab sob sorted hai! lets see opertaion

sns.heatmap(data = data_2)

plt.show()

# yaha dekho   since colorbar value ranges from 1 to 5 let change it as per own wish.
```

```python
# annot once again! 



data_1 = np.linspace(1,10,20).reshape(4,5)  


sns.heatmap(data = data_1, annot = True)   # things are more clear now! 

plt.show()
```

```python
# Magic of annot parameter with fmt :  we can use annot with fmt to change the name of numbers on colors with our desired names 

# Steps to do it:  Create an 2D array(say its name is arr_x) thats adapt with the format of existing 2D array which we used in heatmaqpping
# now pass annot = arr_x  and fmt =" datatype _of elements of arr_x" 
# leets see 




data_i = np.linspace(1,10,10).reshape(2,5)  # 2D array chiye tabhi to  reshape use kiya ! 
print(data_i,"\n\n\n")  # sab sorted hai!


data_f = np.array([["a1","a2","a3","a4","a5"],["b1","b2","b3","b4","b5"]])  # we will use this arr to rename with it the existing arrays name !
print(data_f,"\n\n")


sns.heatmap(data = data_i, annot = data_f, fmt = "s")  # "s"  refers to string data type!
plt.show()

# hurrey ho gya! 
```

```python
# annot_kws = {"fontsize": n ,"color": " "} : This parameter uses to deals with color and fontsize of that charater printed on color patches.



data_1 = np.linspace(1,10,20).reshape(4,5)  


sns.heatmap(data = data_1, annot = True, cbar = False , annot_kws = {"fontsize": 30, "color" : "g"})   # things are more clear now! 

plt.show()

# got it!
```

```python
# cbar parameter: if you want to disable color bar in heatmap interface make it as cbar = False

# bydefault cbar = True means colorbar stays enebled 



data_1 = np.linspace(1,10,20).reshape(4,5)  


sns.heatmap(data = data_1, annot = True, cbar = False )   # things are more clear now! 

plt.show()

# clear !
```

```python
#  .set(xlabel = " ", ylabel = " ")  here its label not labels as    x and y axis are same.
# This is used to givr label name to x and y axis.



iris = sns.load_dataset('iris')  # Flower species data
data_2 = iris.drop(columns = ["species"], axis = 0).head(20)

# print(data_1) # ab sob sorted hai! lets see opertaion

sns.heatmap(data = data_2,vmin = 0, vmax = 10,cmap = "viridis").set(xlabel = "X axis", ylabel = "Y axis")


plt.title("Hello",color = "r")

plt.grid()
plt.show()
```

```python
#   Difference  between count and bar plot :  


# 1 : Counts plots the actual counts of number of record/ that elemts or say entity occcureance by category.

# 2 : Bar plot a number or metric of each category( bydefault bar plot plots mean of variables ( not actual count) by category).
```

```python
# Basic example:


data_tips = sns.load_dataset('tips')  # Restaurant tipping data

print(data_tips,"\n\n\n")  # data is clean!

sns.countplot(x = "total_bill", data = data_tips)
plt.show()

# got it , lets make data some small it is some how bulky.



```
