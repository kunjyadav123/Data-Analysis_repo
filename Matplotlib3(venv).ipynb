#### EXPLORING THE PIE PLOTTING :

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print(pd.__version__)
print(np.__version__)
```

```python
# Firstly explore the need and why we use pie ploting ! Grow some sort of intution.
```

```python
# Syntax of pie plot:

# x = []
# plt.pie()
# plt.show()
```

```python
x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]

plt.pie(x)
plt.show()
```

```python
# Lets add label:  to add multiple lables in pie to each section we use its plural i.e labels.


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]

plt.pie(x,labels = y)
plt.legend() # labels invoke with the label parameter  properly.
plt.show()

# And one  more thing that here colors are shown bydefault but if we want our colour preferences just pass it under color attribute.

```

```python
# color as per our preferneces . but remember here pas it colors not color okay because of multiples .


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
c = ["y","r","b","m","g"]

plt.pie(x,labels = y,colors = c)
plt.show()

```

```python
# explode parameter : Its just used for best showcasing the pie section in bulge out format.
# The more will be explode value the more it will get bulge out , by default it stays 0 and more you increse it for specific part of pie the more it will get buldge out.



x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.0]


plt.pie(x,labels = y,explode = exe)  # Have you seen it!
plt.show()
```

```python
# see it once more!


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe)  # Have you seen it!
plt.show()

```

```python
# autopct = "%0.1f%%"  = this parameter is magical and it will give ariel percentage coverage of each pie section. see


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,autopct = "%0.1f%%") 
plt.show()

# if we use "%0.2f%%"  then it will add one zero at end i.e in place of 60.0%  it will make % as 60.00% simple.
```

```python
# Shadow fn  : Its basically shows shadow of pie interface  just pass shadow = True  and shadow = False id bydefault.


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = True)  # seen it.
plt.show()
```

```python
# radius parameter : Its deals with the radius of pie circle by default it stays radius  = 1.



x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = False,radius = 1.5)  # Got it!
plt.show()
```

```python
# labeldistance parameter : its prints label names at distance you provides here , the more will be its value the more distant the label name will print from pie plot. bydefault its value remains at 1. zero se kam rakhoge to pie ke inside hi print hogi.


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = False,radius = 1.,labeldistance = 1.5)  # Got it!
plt.show()

# But usuallly we dont use it kyu ki label name itna dur sahi nii lgta hai. to vo bydefault hii shai hai 
```

```python
# startangle parameter : its basically the stats the angular inclination with first pie plot  with that angle which you  passed.by default to ye 0 se start hota hai.


# And pies plotiing occurs in anticlockwise oreintation  by default, agar specify kr do to clockwise ho  jayega (counterclock parameter se ) vo baad me dekhgene just wait for that , abhi ke liye startangle samjho aur dekho .



# startangle = 90

x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = False,radius = 1.,labeldistance = 1.5,startangle = 90)  # Got it!
plt.show()
```

```python
# startangle = 0


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = False,radius = 1.,labeldistance = 1.,startangle = 0)  # Got it!
plt.show()
```

```python
# textprops = ({"fontsize" : n})  whwre n is some number. This property deals with ther font size of label,% etc of pie elements. Lets see.




x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = False,radius = 1,labeldistance = 1,startangle = 0,textprops = ({"fontsize" : 20}))  # Got it!
plt.show()

# Jise font zize ho waise hi value pass karo.
```

```python
# Counterclock : By default the pie section name prints anticlockwise , but here we may rotate this factors as per our choice.
# conterclock = True (ye to bydefault hai) and counterclock = False this maintains the clockwise orientation.




x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = False,radius = 1,labeldistance = 1,startangle = 90,textprops = ({"fontsize" : 20}),counterclock = False)  # Got it!
plt.show()
```

```python
# wedgeprops = ({"linewidth" : n}) here n is any number. remember here  it deals with the thickness of edges / shadow of pies .see



x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = True,radius = 1,labeldistance = 1,startangle = 90,textprops = ({"fontsize" : 10}),counterclock =False ,wedgeprops = ({"linewidth" : 10}))  # Got it!
plt.show()
```

```python
# rotatelabels = True , its basically rotate the name of labels of pie, and rotatelabels = False is bydefault.



x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = True,radius = 1,labeldistance = 1,startangle = 90,textprops = ({"fontsize" : 10}),counterclock =False ,wedgeprops = ({"linewidth" : 10}),rotatelabels = True)  # Got it!
plt.show()
```

```python
# Primary Title :


x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = True,radius = .8,labeldistance = 1,startangle = 90,textprops = ({"fontsize" : 10}),counterclock = True ,wedgeprops = ({"linewidth" : 10}),rotatelabels = False)  # Got it!
plt.title("wscube Tech")

plt.show()
```

```python
# showcase the broad view of labels by plt.legend() 



x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = True,radius = .8,labeldistance = 1,startangle = 90,textprops = ({"fontsize" : 10}),counterclock = True ,wedgeprops = ({"linewidth" : 10}),rotatelabels = False)  # Got it!
plt.title("wscube Tech")
plt.legend()  # this give the name of tittle in breif , we may specify the position of titles by loc see
plt.show()
```

```python
# Changing the title position by loc .


# loc = 1 means ttitle position in 1st quadrant
# loc = 2 means ttitle position in 2nd quadrant.
# loc = 3 means ttitle position in 3rd quadrant.
# loc = 4 means ttitle position in 4th quadrant.



x = [10,20,30,40,50]
y = ["python","java","C++","C","Html"]
exe = [0.3,0.0,0.0,0.0,0.2]


plt.pie(x,labels = y,explode = exe,autopct = "%0.1f%%", shadow = True,radius = .8,labeldistance = 1,startangle = 90,textprops = ({"fontsize" : 10}),counterclock = True ,wedgeprops = ({"linewidth" : 10}),rotatelabels = False)  # Got it!
plt.title("wscube Tech")
plt.legend(loc= 1)  
plt.show()
```

#### Exploring the Stem Plot:

```python
# Again before stating it go and explore what is it why we 8se its and  about it applications.
```

```python
# Basic working synatx is simple as .

# x = []
# y = []

# plt.stem(x,y)
# plt.show()
```

```python
# Basics example 


x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y) 
plt.show()



```

```python


x = [1,2,3,4,5,6]
y = [12,10,25,8,20,5]
 
plt.stem(x,y) 
plt.show()
```

```python
# linefmt  parameter is used to depic differnt type of line .see

# we have various type of such lines such as "-" for solid line ,"--" dashed line  etc. by the way solid line is bydefault.


x = [1,2,3,4,5,6]
y = [12,10,25,8,20,5]
 
plt.stem(x,y, linefmt = "-.") 
plt.show()
```

```python
# markerfmt parameyter is is just  basically used for dealing with the marker(that tip ball) texture along with its color.see
# Various markerfmt are .,*,+ ,o(bydefault)etc.


x = [1,2,3,4,5,6]
y = [12,10,25,8,20,5]
 
plt.stem(x,y, linefmt = "-.",markerfmt = "*") 
plt.show()
```

```python
# we can deal with marker color in same markerfmt parametr just pass it as y. , r* etc see




x = [1,2,3,4,5,6]
y = [24,15,25,40,20,18]
 
plt.stem(x,y, linefmt = "-.",markerfmt = "r+") 
plt.show()
```

```python
# bottom parameter : we know bydefault the vertical interval numbersing starts with the 0 but if we use bottom paremeter we may use it to change 0 from othe rdesired number.



x = [1,2,3,4,5,6]
y = [12,10,25,8,20,5]
 
plt.stem(x,y, linefmt = "-.",markerfmt = "r+",bottom = 10) 
plt.show()

# have you seen some thing it may change it diecetion but things are completely fine.
```

```python
# once more!



x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom = 25) 
plt.show()
```

```python
# once again !

# thing are totally justified here!


x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom =75) 
plt.show()
```

```python
# basefmt parameter : Its baically deals with the color of base and its texture means of which type of line we need.


x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "mo",bottom =22,basefmt = "y")  # only color let ow see with linetexture.
plt.show()
```

```python
# once more.


x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom =22,basefmt = "b-.") 
plt.show()
```

```python
# label parameter , it works with the plt.legend()


x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom =15,basefmt  = "m-.",label = "Python !") 
plt.legend()
plt.show()
```

```python
# use_line_collections , it uses to make each line color differnt different.  
# if we use  , use-line_collection  = True , all line color will be same . Thois cas is bydefault.
# if we use , use_line_collection  = False , color of line will be 






x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom =15,basefmt  = "m-.",label = "Python !",use_line_collection = True) 
plt.legend()
plt.show()


# is topic ko hold pe rakho abho ke liye .
```

```python
# orientation : Its basically used for the vertical and horizontal plotiing , for vertical it is bydefault but fo horizonatl pass it .

# We may also pass the labels of each axis and , title to dekh hi liye hai.






x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom =15,basefmt  = "m-.",label = "Python !",orientation = "horizontal")
plt.title("Python!")
plt.xlabel("x axis") 
plt.ylabel("y axis")
plt.legend()
plt.show()
```
