#### Axis Related parameters () :

```python
# xticks   and  yticks  parameter. its basically makes the axis intervals name as which is get passed in it . see



x = [2,3,4,5]
y = [6,7,4,2]
 


plt.plot(x,y,label = "Normal Line Plot")  
plt.legend() 


plt.show()

# see firstly axis lines points without the ticks! now let see with the ticks.
```

```python

x = [2,3,4,5]
y = [6,7,4,2]
x1 = [1,2,3,4,5,6,7]


plt.plot(x,y,label = "Normal Line Plot") 
plt.legend()  

plt.xticks(x)
plt.yticks(x1)
plt.show()


# now see the axis points with the ticks.
```

```python
# we may replce ticks points with labels element see !  and labels  gets invoke with the plt.legend()  so plese remember it!




x = [2,3,4,5]
y = [6,7,4,2]
x1 = [1,2,3,4,5,6,7]


plt.plot(x,y,label = "Normal Line Plot") 

plt.xticks(x,labels = ["Python","C++","C","Java"])  # put labels not label.
plt.yticks(x1)

plt.legend()

plt.show()
```

```python
# plt.xlim()  and plt.ylim() :   Its basically deals with the axis points see how it works


x = [2,3,4,5]
y = [6,7,4,2]
x1 = [1,2,3,4,5,6,7]


plt.plot(x,y,label = "Normal Line Plot") 
plt.legend()  

plt.xlim(0,10)  # its print point from 0 to 10 in x axis .
plt.ylim(1,8)   # its prints point from 1 to 8 in y axis.
plt.show()

# You may see what is happening here!
```

```python
# plt.axis([a,b,c,d])  here a to b deals with the x axis numbetring and c to d deals with the y axis numbering.




x = [2,3,4,5]
y = [6,7,4,2]
x1 = [1,2,3,4,5,6,7]


plt.plot(x,y,label = "Normal Line Plot") 
plt.legend()  
 
plt.axis([1,8,0,9])  # clear hai!
plt.show()
```

#### Working with text in Matplotlib :

```python
# we  know our basic  text depicting methods such as labels, titles etc.


# But 



# If we wish to print any text at any desired corrdinate of plot we uses    plt.text() methods 

# where   plt.text(a,b,"Txt_name")  where  (a,b) shows the  coordinate names and TXt_name is name of text that will print there. see



x = [2,3,4,5]
y = [6,7,4,2]
x1 = [1,2,3,4,5,6,7]


plt.plot(x,y,label = "Normal Line Plot") 
plt.legend()
plt.axis([0,10,1,10])
plt.text(8,7,"Txt_Python")
plt.show()
```

```python
# once more!

# we can deals with the fontsize of that text.
# also we can give style to that text.

x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom = 25) 
plt.text(3,10,"Java_3,10",fontsize = 20,style = "italic")
plt.show()
```

```python
# bbox = {"facecolor" = "colr_name" }  Method. This methods basically adds colors around that text.


x = [1,2,3,4,5,6]
y = [25,60,40,30,20,5]
 
plt.stem(x,y , markerfmt = "yo",bottom = 25) 
plt.text(3,10,"Java_3,10",fontsize = 20,style = "italic",bbox = {"facecolor" : "r"})  # got the bbox .
plt.show()
```
