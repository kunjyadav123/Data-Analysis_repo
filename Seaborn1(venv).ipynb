```python
# simple  example.

data_tips = sns.load_dataset('tips')  

sns.barplot(x = "sex",y = "total_bill",data = data_tips)

plt.show()

# Have you seen labesl are prebulid and shown  as even without laleb and legend() intact unlike in case of matplotlib.
```

```python

# hue = " columns/group/category_name "   parameter , This parametre is use to do categorical differntiation based on assigning different categories and different color.
# this helps to understand and identifying paterns , relationship between various groups and categories.
# color of huw will be show by different color for better understanding of patterns , relationship anomgs groups. 
# again I warn you that the put category, column , group name should be from same dataframe that we used . 


data_tips = sns.load_dataset('tips')  

sns.barplot(x = "sex",y = "total_bill",data = data_tips,hue = "smoker")

plt.show()
```

```python
# order parameter: its basically deals with the written position of x axis elemnt ! jis order me pass karoge whi hoi jayga  
# for example as here order is   male then female .let change it writtten order.



data_tips = sns.load_dataset('tips')  

sns.barplot(x = "sex",y = "total_bill",data = data_tips,hue = "smoker",order = ["Female","Male"])  # oredr got changed.

plt.show()
```

```python
# hue_order paramter :  This parameter used to change order of that multiple elemnts of a barplot.
# as here order is yes (smoker), then no(smoker) let change it 



data_tips = sns.load_dataset('tips')  

sns.barplot(x = "sex",y = "total_bill",data = data_tips,hue = "smoker",order = ["Female","Male"],hue_order = ["No","Yes"])  # bars elemnts order gets changed.

plt.show()

```

```python
# Have you notices with less parameters we seen and deals with multiple barplots in seaborn unlike matplotlib where we requires some others parameters loke subplots for this purpose . This is main advantaage of seaborn over matplotlib.
```

```python
# ci parameter : notice at notch of bars we have that small straight stick ,we can uses such sticj=k for showing tthe intervals 
# and ci parameter uses to change the length  of that stick. see

# Value of ci must be between 0 to 100 (100 is exclusive).

# use it as errorbar = ("ci",val)

# ci maens confedential interval.





data_tips = sns.load_dataset('tips')  

sns.barplot(x = "sex",y = "total_bill",data = data_tips,hue = "smoker",order = ["Female","Male"],hue_order = ["No","Yes"],errorbar  = ("ci", 80))  

plt.show()
```

```python

#  orient parameter : its basically deals with the vertical or horizontal bars.

# bydefault orient parameter is "v" i.e vertical and we can use it in all condition as
# but if we want to make orient = "h" i.e horizontal make sure first that your x and y both are of numerical datas  nii to error ayega. next cell me dekho.




data_tips = sns.load_dataset('tips')  

sns.barplot(x = "sex",y = "total_bill",data = data_tips,hue = "smoker",order = ["Female","Male"],hue_order = ["No","Yes"],errorbar  = ("ci", 80),orient = "v" )  

plt.show()
```

```python
# orient = "h"  see basic idea mil gya hoga.


data_tips = sns.load_dataset('tips')  

sns.barplot(x = "size",y = "total_bill",data = data_tips,hue = "smoker",errorbar  = ("ci", 80),orient = "h" )  

plt.show()
```

```python
# saturation parameter = its deals with the color boldness of bar see higher the value higher the color boldness and vice versa.





data_tips = sns.load_dataset('tips')  

sns.barplot(x ="sex",y = "total_bill",data = data_tips,hue = "smoker",errorbar  = ("ci", 80),palette = "viridis" , saturation = 1)  

plt.show()

```

```python

# color and width parameter of ci stick 

# err_kws = {"color": "r", "linewidth" : n}  use this to deal with such attributes of ci stick



data_tips = sns.load_dataset('tips')  

sns.barplot(x ="sex",y = "total_bill",data = data_tips,hue = "smoker",errorbar  = ("ci", 80),palette = "Accent",err_kws = {"color": "r", "linewidth" : 2} )  

plt.show()
```

```python


# once more!


# err_kws = {"color": "r", "linewidth" : n}


data_tips = sns.load_dataset('tips')  

sns.barplot(x ="sex",y = "total_bill",data = data_tips,hue = "smoker",errorbar  = ("ci", 80),palette = "Accent",err_kws = {"color": "r", "linewidth" : 8} )  

plt.show()





```

```python
# alpha parameter :  Its basically deals with the intesity of color in interface.

# value of alpha parameter must be between 0 to 1.

data_tips = sns.load_dataset('tips')  

sns.barplot(x ="sex",y = "total_bill",data = data_tips,hue = "smoker",errorbar  = ("ci", 80),palette = "Accent",err_kws = {"color": "r", "linewidth" : 8}, alpha = 0.4 )  

plt.show()

```

```python
# sns.set(style = "  ")  parameter : this syntax is used to deal with the background orientation of inetface.
# we may pass here ,,    dark , darkgrid etc. see 




data_tips = sns.load_dataset('tips')  
sns.set(style = "darkgrid")
sns.barplot(x ="sex",y = "total_bill",data = data_tips,hue = "smoker",errorbar  = ("ci", 80),palette = "Accent",err_kws = {"color": "r", "linewidth" : 8}, alpha = 0.4 )  

plt.show()
```
