```python
# As per our usages we may either have same colour or different colours for different - differents dots in scatteres plot.

               # Same colour ! in every dot.


y = [1,4,6,3,9]
x = ["a","b","c","d","e"]

plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = "r",label= "Coloured")    
plt.legend()            
plt.show()
```

```python
# Different Different  colours for Different dots .

y = [1,4,6,3,9]
x = ["a","b","c","d","e"]

c = ["r","y","b","g","m"]  
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = c)               
plt.show()
```

```python
# Sizes parameter : It basically deals with the sizes of dots which you sees in scatter plotting.


y = [1,4,6,3,9]
x = ["a","b","c","d","e"]


s = [5,100,40,20,30] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
c = ["r","y","b","g","m"]    # colors attributes.
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = c,sizes = s) # sizes attribute !
plt.show()

```

```python
# Alpha Parameter : Its basically deals with the intesity of colours in scaterr dot.
# Value of Alpaha paremeter must be between 0 to 1 and again more the value of alpha more will be its intensity but value will be between 0 to 1 and by default alpha parametre value ramains 1 i.e when no aplha parameter does not get passed.


y = [1,4,6,3,9]
x = ["a","b","c","d","e"]


s = [5,100,40,20,30] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
c = ["r","y","b","g","m"]    # colors attributes.
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = c,sizes = s,alpha = 0.6) # Have you seen colour gets fade!
plt.show()

```

```python
# let see one more example where alpha  = 0.2 so as value of alpha is less so its intensity will be more less.


y = [1,4,6,3,9]
x = ["a","b","c","d","e"]


s = [5,100,40,20,30] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
c = ["r","y","b","g","m"]    # colors attributes.
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = c,sizes = s,alpha = 0.2) # Have you seen colour gets more fader now coz here valur is much lesser.
plt.show()
```

```python
# Marker parameter : Its changes Texture/Design of that dots into other texture.

# For star shaped marker use Mraker = "*" okay. lets see now.



y = [1,4,6,3,9]
x = ["a","b","c","d","e"]


s = [50,100,40,80,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
c = ["r","y","b","g","m"]    # colors attributes.
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = c,sizes = s,marker = "*") # sizes attribute !
plt.show()
```

```python
# Dealaing with the bolundaries of that dots/ stars: 

# 1 : We may change the edge colour of boundaries of dots/edges via the edgecolour parameter.
# 2 : We may  change the edge/boundaries thickness of the dots or stars by linewidth parameter. Lets see



y = [1,4,6,3,9]
x = ["a","b","c","d","e"]
sizes = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
color = ["r","y","b","g","m"]    # colors attributes.
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,c = color,s = sizes,marker = "*",edgecolor = "m") # Have you see the edge / boundary colour of dots/ stars changed.
plt.show()
```

```python
# Lets now change the thickness of that edgecoloured line by linewidth .More the edgewidth value more will be the the thickness of edgeline and vice versa.
# See and one things the value of edgewidth may be out of 0 to 1. so feel free okay.

y = [1,4,6,3,9]
x = ["a","b","c","d","e"]
 
s = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
c = ["r","y","b","g","m"]    # colors attributes.
plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,color = c,sizes = s,marker = "*",edgecolor ="m",linewidth = 2) # Have you see the edge / boundary colour of dots/ stars changed.
plt.show()
```

```python
# cmap attribute: Its basically works on numeriacal to color relation . Actulay we have multiple inbulit colou codes in which multiple colours exist in one codes words and each colour codes have its own code numbers and numbers fluctuate between 1 to 100. lets see how it works.
# See one thing that tjis attribute use for manipulating the colour 0f dots/stars. 
# And we have multiple colour codes and in that codes we have multiple colours and theor colours are specifed by the numbers between 0 to 100 and each colour in that code have different numbers.


# Imp Note : Ek baat when you use cmap property and then while giving the code no of colors make sure you stores in variable name as  color  dont use c .
# And within  plt.scatter() use it as c= color not color = c . Plese remember this synatx otehtwise you will face error.
# lets see how it works.


y = [1,4,6,3,9]
x = ["a","b","c","d","e"]


s = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
color = [20,50,60,90,9]  


plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y,c = color,sizes = s,cmap = "viridis") # Here "viridis" is prebuilt codes name . and we have multiple such coedes name which inclueds sets  of 100 colours and its each different colour is defined by different code from 0 to 100.
plt.show()
```

```python
# But how will be know what is color code for which colour in a color code. Lets see
# to get it we simmpy call plt.colorbar() and it will give you a guide .


# But before calling the plt.color(), do following things 
# Donts put color variable  name bu single letter instaed of c put color
# and then in plt.scatter() put following   as c= color not color = c and copy this format otherwise you will face error.
# so keep such rules in mind before using the cmap attribute.
y = [1,4,6,3,9]
x = ["a","b","c","d","e"]

s = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
color = [20,50,60,90,9]  


plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y, c = color,sizes = s,cmap = "viridis")
plt.colorbar()
plt.show()
```

```python
# We may change the intensity of scatter graph via alpha attribute method see 



x = ["a","b","c","d","e"]
y = [1,4,6,3,9]

s = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
color = [20,50,60,90,9]  


plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y, c = color,sizes = s,cmap = "viridis",alpha = 0.2)  # low valus of alpha makes the scatter plot faded.
plt.colorbar()
plt.show()
```

```python
# Adding label within the sactter interface of colourchart. you cannot add directly label in scatter plots colourchart here label adds in only at side of colorbar and we call it as following way.
# we will do it via  put   t = plt.colorbar()
#                          t. set_label("Label name")

# Aur waise agar simple scateer interface ke inbetween lagana hai to use   label = "Label_name " in plt.plot() and put  plt.legend()

y = [1,4,6,3,9]
x = ["a","b","c","d","e"]

s = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
color = [20,50,60,90,9]  


plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y, c = color,sizes = s,cmap = "viridis",label = "Dots/Stars")
t = plt.colorbar()
t.set_label("Colorfulllll World", fontsize = 12)  # and we may change label's font size in such a way.4
plt.legend()
plt.show()

# Thogs are little bit cmolex so practice them intensly.
```

```python
# We may pass multiple scatter plots within the same plot of scatter ! see the way.




x = ["a","b","c","d","e"]
y = [1,4,6,3,9]
z = [2,9,10,5,3]

s = [300,500,90,220,300] # Sizes attribute. More the size numbers more the size sof dot and vice versa we may see same here !
color = [20,50,60,90,9]  


plt.title("Scatter Plot",fontsize = 30)  # Graph tittle.
plt.xlabel("Letter",fontsize = 20)   # X label.
plt.ylabel("No.",fontsize = 15) # Y label.

plt.scatter(x,y, c = color,sizes = s,cmap = "viridis") # This follows the cmap attribute.
plt.scatter(x,z, color = "r",sizes = s)   # This is plain one.  Okay

t = plt.colorbar()
t.set_label("Colorfulllll World", fontsize = 12)  # and we may change label's font size in such a way.

plt.show()
```

```python
# Edge Color  attribute : Its basically gives edges color  to the hist bars . siumple




x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]
plt.hist(x,color = "r",edgecolor = "y")
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()
```

```python
# bins parameter = Its basically maintains the specific range in gappping the hsit bars intervals.see
# We give bins value as parameter as per our need and requirement.


# Dessired gapping mil gyiii na ! yehh
# Aur agar bins parameter use nii karoge to matplotlib apne acccording intervals decide karega.
```

```python
# Range parameter : Its also give intervals(its works based on broad range not on specific interval) but its need just two argument in tuples(a,b)  whwre a is starting no of d=first interval and b is last.
# Remember one thing before using the range parameter make sure you do these 3 task
# 1 : put bins parameter on auto  as  "auto" just no value passage.
# 2 : put the range as (a,b) where a, b are numbers. and
# follow this syntax : in plt,hist   code as     "auto",(a,b)  such that (a,b)  should be the 3rd  argument within the plt.hist() 
# Exaple 


x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

plt.hist(x,"auto",(0,200),color = "r",edgecolor = "y")
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()
```

```python
# Cumulative parameter : Its basically ahve two paraameter 1 and -1 
# for 1 its arranges hist in ascending ordr and in -1 it arranges hist in descending order.

# cumulative  = -1


x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "y",cumulative = -1)
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()
```

```python

# cumulative  = 1


x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "y",cumulative = 1)
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()
```

```python
# Bottom parameter : Usaually by default the vertical interval  starts with 0 but if you want to start it by other value we may use bottom parameter for this purpose. see



x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "y",bottom = 10)
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()


# Now here vertical interval statts with the 10.
```

```python
# log attribute / papeameter : its absically make the vertical frequency valus to depict in logirathmic function.
# To ebable it just pass    log = True and by default it is log = False
# see now 



x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "m",log = True )
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()
# Dont worry everthing is sorted.
```

```python
# label parameter : we know it it gives label to hist inetface and it works with the plt.legend().


x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "m",log = False ,label = "python")
plt.legend()
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.show()
```

```python
# plt.axvline()    : this is very important function to invoke a vertical line in histbars and it can we used for various indication purpose .
# thos line basically align to specific valus which you provides it. we may give color to taht line via color parameter.


x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "m",log = False ,label = "python")
plt.legend()
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.axvline(65,color = "g")
plt.show()
```

```python
# We may give laleb to that plt.axvline()  just pass label parameter and that plt.legend( )  is must   for that label.
# each alabel requires it own plt.legend( ), sabke liye apna apna alag alag but plt.legend( ) should come after the label name .

x = [12,15,5,70,20,45,60,38,40,60,55,75,80,27,10,45,30]

k = [10,20,30,40,50,60,70,80,90]
plt.hist(x,color = "r",bins = k,edgecolor = "y",log = False ,label = "python")
plt.legend()
plt.title("Histogram Plots")
plt.xlabel("x Axis")
plt.ylabel("No")
plt.axvline(65,color = "m", label  = "mysterious line")
plt.legend()
plt.show()
```
