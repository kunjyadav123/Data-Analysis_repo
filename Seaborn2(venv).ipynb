```python
# once more

data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data
print(penguins,"\n\n\n")


sns.histplot(data_penguins["body_mass_g"])
plt.show()
```

```python
# bins parameter: it basically dels with the organosed horizontal interval as pr need .
# as you may see here interval are unorganised , but if we want that intervals as per our choice.
# then hist plots wiiiil be shown in particular intervals(we have already seen bins parameter in matplotlib already).
# due to bins we may understand data more effectively  as intervalwise. 

data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data

sns.histplot(data_penguins["body_mass_g"], bins = [1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000,6500,7000])
plt.show()
```

```python
# kde(kernel density operator) parameter : Its basically plots a curve along our hist. 
# to enable it pass kde = True and and bydefault it is False.
# to use kde bins ka hona jaruri nii hai , hai to theek ni to bhii koi problem nii.
# now go and explore what is purpose of that kde.

data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data

sns.histplot(data_penguins["body_mass_g"], bins = [1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000,6500,7000], kde = True)
plt.show()
```

```python
# rug parameter : Its basically deals with the envoking the kernel density of x axis .
# to enable it pass  rug = True   and bydefault it is False 


data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data

sns.histplot(data_penguins["body_mass_g"],kde = True, rug = True )
plt.show()

# kuch error aa gya hai we will sort it later .
```

```python
# color parameter : known 




data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data

sns.histplot(data_penguins["body_mass_g"],kde = True, color = "y" )
plt.show()
```

```python
# log_scale parameter : its basicalyy makes interval data into log table.
# juts pass log_scale = True , bydefault its stays False.



data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data

sns.histplot(data_penguins["body_mass_g"],kde = True, color = "y" , log_scale = True)  # see there
plt.show()
```

```python
# plt.grid()
# plt.title()  are well knwsn to us .
# as we have multiple prameters in those two parameter as we already seen them before so we will just see only without any sub pramter.



data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data

sns.histplot(data_penguins["body_mass_g"],kde = True, color = "y" )
plt.grid()
plt.title("My Histogram Plot !")
plt.show()
```

```python
# Datasets / Dtaframes(Seborn comes up with several built in dataset that can we load easily.)


import seaborn as sns

# here sns.load_dataset("data")

# Load example datasets
tips = sns.load_dataset('tips')  # Restaurant tipping data
iris = sns.load_dataset('iris')  # Flower species data
penguins = sns.load_dataset('penguins')  # Penguin species data
flights = sns.load_dataset('flights')  # Flight passenger trends
diamonds = sns.load_dataset('diamonds')  # Diamond prices and features

# To make that dataframe print(access) first time we need internet connection , abee file aise ho thodi acces hp jaygi first time .So be carefrl.

# These datasets are prebulit and accessed from gitrepository so practice hard.
# we have also some downloades dataframes in system we can also use them when require, also we can create an instat dataframe for practicing whenever required.
```

```python
# Basic working syntax of scatter plots.



# data_name =  any dataframe or any dataset i.e csv, xlsx, json  or use as  data_a = sns.load_dataset("data_name ") from github etc aor from any source.
# sns.scatterplot(x= " ",y = " ",data= data_name)  # passing data is optional ! but as seaborn is known for its clasic plotting so we passes data frames here! usually
# plt.show()


# here  x and y are datas are must from dataframe that we passed in as data = dataframe.
# here data must be of dataframe(of pandas) okay not list.


# Here x ,y and data defines core structure . remember x any are prebulit and they defines the axes orientataion.
# We specify dataframe name  within sns.lineplot ()  because so

```

```python
# data to sorted hai!


data_penguins = sns.load_dataset('penguins')  # Restaurant tipping data
print(penguins,"\n\n\n")




sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins)
plt.show()
```

```python
# hue = " columns/group/category_name "   parameter , This parametre is use to do categorical differntiation based on assigning different categories and different color.
# this helps to understand and identifying paterns , relationship between various groups and categories.
# color of huw will be show by different color for better understanding of patterns , relationship anomgs groups. 
# again I warn you that the put category, column , group name should be from same dataframe that we used . 


sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex")
plt.show()
```

```python
# size and sizes :  we know it lets use it now.
# sixe ye batayega ki kis gropup/ columnn ke sizes sedeal karna hai
# sizes yes batayega ki jitne varity hai size group ke attribute me unke rspective size kya hoga.
# as here for men  = 50 and female = 15.
# size ke parameter(column ke jitne variety hongi)   , unte hi vaules sizes me pass hongi.

sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex",size = "sex", sizes = (15,50))
plt.show()
```

```python
# palette parameter : We know it ! 



sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex",size = "sex", sizes = (15,50), palette = "Accent")
plt.show()
```

```python
# palette once more!



sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex",size = "sex", sizes = (15,50), palette = "viridis")
plt.show()
```

```python
# style  parameter: it basuically will deals with the texture of scatter . by its own 
# just pass style  = "colmn / group name"
# see

sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex",size = "sex", sizes = (20,50), style = "sex")
plt.show()
```

```python
# alpha parameter : Its deals with the color intesity of interface.

# Its value must be betwen 0 to 1 and bydefault it stays 1.


sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex",size = "sex", sizes = (15,50), style= "sex", alpha = 0.4)
plt.show()


# difffrence is clear.
```

```python
# markers (not marker) parameter : Its basically deals with the scatter texture.
# pass value dictionary.


sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins, hue = "sex",size = "sex", sizes = (30,50), style= "sex", alpha = 1, markers = {"Male":"*", "Female": "o"})
plt.show()
```

```python
# dealing with limited data 
# use head and tail .

sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm",  data = data_penguins.head(20), hue = "sex",size = "sex", sizes = (15,50), style= "sex", alpha = 1)
plt.show()
```
