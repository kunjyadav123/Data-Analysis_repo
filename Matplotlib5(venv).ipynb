#### Exploring the fill_between plot:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print(pd.__version__)
print(np.__version__)
```

```python
# lets goo. 


x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.title("Fill_between Plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
# color parameter : it works on color of that filled portion.


x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.title("Fill_between Plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y)
plt.fill_between(x,y,color = "m")
plt.show()
```

```python
# where parameter: Its basically used to give some sorted condition underwhich the condition fullfilled portion fills.
# see how it pass and  it syntax.




# But remember ye where function normal list pe work nii karta ahi , for that purpose we need  array( ohh we can convert the normal list into array via using numpy).

x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.title("Fill_between Plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y)
plt.fill_between(x,y,color = "m" ,where =(x>=2) & (x<=4))   # error aya just bcoz of list , now let see how it works on the  array , it will work surley on array 
plt.show()
```

```python
   # Now it works as x and y are array!



x = np.array([1,2,3,4,5,6])
y = np.array([1,2,3,4,5,6])
plt.title("Fill_between Plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y)
plt.fill_between(x,y,color = "m" ,where =(x>=2) & (x<=4))   # err or aya just bcoz of list , now let see how it works on the  array , it will work surley on array 
plt.show()
```

```python
# once more ! 



x = np.array([1,2,3,4,5,6])
y = np.array([1,2,3,4,5,6])
plt.title("Fill_between Plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y)
plt.fill_between(x,y,color = "y" ,where =(x>=3) & (x<=4))   # err or aya just bcoz of list , now let see how it works on the  array , it will work surley on array 
plt.show()
```

```python
# alpha parameter : It deals with the intensity of filled color ! 
# alpha value must be between 0 to 1 , bydefault it stays on 1.


x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.title("Fill_between Plots")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y)
plt.fill_between(x,y,color = "m",alpha  = 0.5)
plt.show()
```

#### Exploring the Sub-Plots :

```python
# Its syntax is exciting!
# In ecosystem of subplots we may plot multiple subplots of same type and either plot subplots of different types.


# x = []
# y = []
# plt.subplot(no_rows,no_cols,index_no_subplot)   here (no_rows , no_cols)  are that dimension in which you want subplot . for example (2,2) means that the you have 4 index value and we may plot 4 subplot, same (2,1) shows that we have 2 index so we can plot 2 subplots.
# plt.plot/bar/pie()  # jis  bhi type ka plot ho depict kr do .
# plt.show()
```

```python
# plotting sub plot of same type.



x = [1,2,3,4,5,6]
y = [6,3,4,8,2,9]

plt.subplot(1,2,1)
plt.plot(x,y)


x = [1,2,3,4,5]
y = [5,9,2,10,12]

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()

# Got the fundamental ! Here each subplot is independent plot in itself 
# we may add features , methods , parameter in desired subplots as per requirement. letsee again .
```

```python
# to add title to respective subplot use plt.title , but to make title of entire subplot use    plt.suptitle()    here  it is suptitle  not subtitle .






x = [1,2,3,4,5,6]
y = [6,3,4,8,2,9]
plt.subplot(2,2,1)
plt.title("Subplot1")
plt.ylabel("y", color = "r")
plt.plot(x,y)




x = [1,2,3,4,5]
y = [5,9,2,10,12]
plt.subplot(2,2,2)
plt.title("Subplot2")
plt.plot(x,y)





x = [1,2,3,4]
y = [9,5,4,1]
plt.subplot(2,2,3)
plt.title("Subplot3",color = "m")
plt.plot(x,y)




x = [1,2,3,4,5,6,7]
y = [8,3,4,1,3,6,5]
plt.subplot(2,2,4)
plt.title("subplot 4", fontsize = 15)
plt.plot(x,y)



# specify plt.title afyter specifing plt.subplot(2,2,n) because it will amke print title at proper place as we defined position before it via plt.plot(2,2,n)

plt.suptitle('Collection of Subplot!')

plt.show()
```

```python
# Now let see different types of subplots.
# we will see here lot od paremeter, attribute , methods in respective subplots stay focused how things actually works.



x = [1,2,3,4,5,6]
y = [6,3,4,8,2,9]
plt.subplot(2,2,1)
plt.title("Line")
plt.ylabel("y", color = "r")
plt.plot(x,y)




x = [1,2,3,4,5]
y = [5,9,2,10,12]
plt.subplot(2,2,2)
plt.title("Bar")
plt.bar(x,y)
plt.grid()





x = [10,25,32,65]
ex = [0.3,0.0,0.0,0.0]
plt.subplot(2,2,3)
plt.title("Pie",color = "m")
plt.pie(x,autopct = "%0.1f%%",explode = ex)






x = [1,2,3,4,5,6,7]
y = [8,3,4,1,3,6,5]
plt.subplot(2,2,4)
plt.title("Stem", fontsize = 15)
plt.stem(x,y, markerfmt= "*",label  = "stemwale ! ")
plt.legend()


# specify plt.title afyter specifing plt.subplot(2,2,n) because it will amke print title at proper place as we defined position before it via plt.plot(2,2,n)

plt.suptitle("Collection of Subplot!", color  = "b")

plt.show()
```
