```python
# lets solve that bar overlapping problem here! Be patient it is little complex so plese.

# Here we will see the multiple bar graph of different subjects in same bar inferface and more important it will be non overlapping. But stay patient.

# Its include varoos steps see precsesly.






# step 1 : A new array creation from numpy . And  considering that arry in bar creation insted of x . see it heree.


x = ["Pankaj Yadav","Anup pandit","Abhinav Patel","Nikhil singh"]
y = [25,50,100,80]
z = [80,60,90,40]
p = np.arange(len(x))    # We creaeted a new array 

width = 0.2   # Loweing index so that bars get easily adjust.
plt.bar(p,y,width,color = "r",label = "Popularity of Language(1)")   # using p in palce of x.
plt.bar(p,z,width,color = "y",label = "popularity of language(2)")   # using p in place of x.
plt.show()


# Ek cheez different hui horizontal axis pe p ki values print hui hai.
# But we want non overlapppng bars and with languages name(x) as horizontal axis.
# just be patirnt and see next step.


```

```python
# step 2 :   Creating another list for non overlapping .se wisely nd try to understan what is raellly going.







x = ["Pankaj Yadav","Anup pandit","Abhinav Patel","Nikhil singh"]
y = [25,50,100,80]
z = [80,60,90,10]
width = 0.2   # Loweing index so that bars get easily adjust.
p = np.arange(len(x))    # We creaeted a new array 

p1 = [ j + width for j in p]

plt.bar(p,y,width,color = "r",label = "Popularity of Language(1)")     
plt.bar(p1,z,width,color = "y",label = "popularity of language(2)")    
plt.show()

# Let in next step we will see the how to put x names in horizontal axis.
```

```python

# Step 3 : Putting x axis in horizontal axis by using the ticks method.  

# Ticks method basically copies the value(which is passed) in that axis . eg xticks(val) so here it will name then x axis by it values.



x = ["Pankaj Yadav","Anup pandit","Abhinav Patel","Nikhil singh"]
y = [25,50,100,80]
z = [80,60,90,40]
width = 0.2   # Loweing index so that bars get easily adjust.
p = np.arange(len(x))    # We creaeted a new array 

p1 = [j + width for j in p]

plt.bar(p,y,width,color = "r",label = "Popularity of Language(1)")     # using p in palce of x.
plt.bar(p1,z,width,color = "y",label = "popularity of language(2)")    # using p1 in place of x.

plt.xticks(p+ width,x)  # This is syntax to use it so be aware.
plt.show()
```

```python
# If you want to rotate the name of horizontal see and use Rotate method.Its rotates by that degree of inclination which you passed.




x = ["Pankaj Yadav","Anup pandit","Abhinav Patel","Nikhil singh"]
y = [25,50,100,80]
z = [80,60,90,40]
width = 0.2   # Loweing index so that bars get easily adjust.
p = np.arange(len(x))    # We creaeted a new array 

p1 = [j + width for j in p]

plt.bar(p,y,width,color = "r",label = "Popularity of Language(1)")     # using p in palce of x.
plt.bar(p1,z,width,color = "y",label = "Popularity of language(2)")    # using p1 in place of x.

plt.xticks(p+ width/2,x,rotation = 60)  # This is syntax to use it so be aware. to keep the names of horizontal axis in middle we simply make it half as you see.
plt.show()

# Note : We cannot use rotation in normal caess it can be used in ticks cases only.
```
