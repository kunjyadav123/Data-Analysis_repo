```python
#simple one :


x = [10,20,30,40,50,60,70,80,90]  # here 10 is depicetd as lowest value by stick and 90 is by highest value by stick.
plt.boxplot(x)
plt.show()
```

```python
# Notch plot : to enable notch plot pass , notch = True and if we not want it pass notch = False ( it is by default).


x = [10,20,30,40,50,60,70,80,90] 
plt.boxplot(x , notch = True) # boxplot is  now notch plot! 
plt.show()
```

```python
# vert property : is basicalyy decide whethert the boxplot will be vertical or horizontal.
# for vert = True it is vertical but for vert = False it remains horizonal, remember that vert = True is bydefault.



x = [10,20,30,40,50,60,70,80,90] 
plt.boxplot(x , notch = True,vert = False) 
plt.show()
```

```python
# xlabel and ylabel.  and title 



x = [10,20,30,40,50,60,70,80,90] 
plt.title("Boxplot",color ="r")
plt.xlabel("X axis", fontsize = 15)
plt.ylabel("Y axis")

plt.boxplot(x , widths  = 0.2) 
plt.show()       
```

```python
# tick_labels parameter : here in boxplot labels ( not tick_label coz of usually multiple labels) we uses it get name of hozizontal axis elements like in place of 1,2 etc ( as here we may name it in place of 1 ) and to replace 1 or any horizontal axis elemnt  pass tick_labels  = ["labels_name "] in plt.boxplot().
# here pass label value within the [] okay.



x = [10,20,30,40,50,60,70,80,90] 
plt.boxplot(x , widths  = 0.2,tick_labels = ["Python"])  # we will
plt.show()       
```

```python
# patch_artist  , its basically use to fill the box with color. 
# to fill box with color pass patch_artist = True and bydefalut the box is empty becaus eby default patch_artist ramains False.





x = [10,20,30,40,50,60,70,80,90] 
plt.boxplot(x , widths  = 0.2,tick_labels = ["Python"],patch_artist = True,)  # we will
plt.show()       
```

```python
# showmean method  : we know median point is shown by an line but this method sjhows us mean value of list.
# to enable it just pass showmean = True


x = [10,20,30,40,50,60,70,80,110] 
plt.boxplot(x , widths  = 0.2,tick_labels = ["Python"],showmeans = True)   # means is slightly above to median here.
plt.show()       
```

```python
# some  ; see their syntax it is mandatory to follow it properly.


# 1 : boxprops = dict(color = "r")  # its basically deals with the color of box.
# 2 : capprops = dict(color = "r")  # it basiallys deals with the color of line denoting the highest and lowest value of list.
# 3 : whiskerprops = dict(color = "r")  # its basically  deals with the color of whisker line.



x = [10,20,30,40,50,60,70,80,110] 
plt.boxplot(x , widths  = 0.2,tick_labels = ["Python"],showmeans = True,boxprops = dict(color = "g"),capprops = dict(color = "m"),whiskerprops = dict(color = "y")),    # means is slightly above to median here.
plt.show()       

# Now everything is sorted.

```

```python
# Plotting the multiple box plot.
# Here we  plots various boxplots in one interface  so do it , we firstly pass various(as per need) list inside a variable  and the we use that single varialbe for that puropose .see
# we follow above approach becoz if we try to plot boxes by its own list name and in its own  their own plt.boxoplot() differently  itwill make print all boxes at same position (boxes will concides to each other ) and this will create a choas.   


x = [10,20,30,40,50,60,70,80,110] 
x1 = [12,24,36,50,65,70,90,100,105]

y = [x,x1] # pass both list value in a list.
plt.boxplot(y , widths  = 0.2,tick_labels = ["Python","C++"],showmeans = True,boxprops = dict(color = "b"),capprops = dict(color = "m"),whiskerprops = dict(color = "y")),    # means is slightly above to median here.
plt.show()   

# Now everything is sorted.
```

```python
# ONCE MORE! 
# tick_label  numbers must be comparitbel with the number of list passed in y variable.


x = [10,20,30,40,50,60,70,80,110] 
x1 = [12,24,36,50,65,70,90,100,105]
x2 = [5,14,34,50,80]

y = [x,x1,x2] # pass both list value in a list.
plt.boxplot(y , widths  = 0.2,tick_labels = ["Python","C++","Java"],showmeans = True,boxprops = dict(color = "b"),capprops = dict(color = "m"),whiskerprops = dict(color = "y")),    # means is slightly above to median here.
plt.show()   
```

#### Exploring the Stack Plot X Aerial Plot :

```python
# Again explore the fundamental and basics of what is it really and what is it applications.
```

```python
# Basic syntax : in short this plot graph deals with the area related plotting.

# x = []
# y = []  where y is area data  usually.

# plt.stackplot(x,y)   # but as per need we may pass more values as list as stack.
# plt.Show()
```

```python
# Basic example :
# Try to grab the crux.
x = [1,2,3,4,5]

area = [2,3,5,4,3]

plt.stackplot(x,area)
plt.show()
```

```python
# Dealing with multiple stack:


x = [1,2,3,4,5]

area = [2,3,5,4,3]
area1 = [4,3,2,1,2]
area2 = [1,3,2,4,3]


plt.stackplot(x,area,area1,area2)  # see and figure out how things works and by the way it is easy to get.
plt.show()

# unitwise figures(aerial figures )  ko dekho it will be more understanding.
```

```python
# labels (not label because here we have multiple stack), parameter to give labale , see to invoke it pass plt.legend()  see now.

# numbers of label must be comparitleb with the numebr of stacks.

# we may pass loc  = n (any interger)  to deal with labels name location. it should be pass within the plt.legend(loc = 2)

x = [1,2,3,4,5]

area = [2,3,5,4,3]
area1 = [4,3,2,1,2]
area2 = [1,3,2,4,3]

l = ["area","area1","area2"]
plt.stackplot(x,area,area1,area2,labels = l)  # see and figure out how things works and by the way it is easy to get.
plt.legend( loc = 2)
plt.show()
```

```python
# colors (yes colors not color) parameter , and one thing numbes of colors atttributes must be compatible with the stacks no .plt




x = [1,2,3,4,5]

area = [2,3,5,4,3]
area1 = [4,3,2,1,2]
area2 = [1,3,2,4,3]

l = ["area","area1","area2"]
plt.stackplot(x,area,area1,area2,labels = l,colors  = ["r","y","m"])  # see and figure out how things works and by the way it is easy to get.
plt.legend(loc = 1)
plt.show()
```

```python
# we may also pass heree label,title and grid parameter.see we may also see various mathod within all such grid , title , label   but we will not see them here we have alredy know thm as we learnt them in previois topics .  



x = [1,2,3,4,5]

area = [2,3,5,4,3]
area1 = [4,3,2,1,2]
area2 = [1,3,2,4,3]
plt.title("Pyhton")
plt.xlabel(" X axis")
plt.ylabel(' Y axis')


l = ["area","area1","area2"]
plt.stackplot(x,area,area1,area2,labels = l,colors  = ["r","y","m"]) 

plt.legend(loc = 1)
plt.grid()
plt.show()
```

```python
# lets see wahi same parameter.

x = [1,2,3,4,5]

y = [4,3,6,7,3]
plt.title("Step plot", color = "m") 
plt.xlabel(" X axis",)
plt.ylabel(" Y axis",fontsize  = 10)
plt.step(x,y)
plt.grid()  # grid () also have its own some methods but we have alredy seen them in our previous plots  so we wiill not learn them here.
plt.show()
```

```python
# marker parameter : Its basically deals with the the symbols used at corner of step.
# color parameter : It uses to deal with the color of step.


x = [1,2,3,4,5]

y = [4,3,6,7,3]
plt.title("Step plot", color = "m") 
plt.xlabel(" X axis",)
plt.ylabel(" Y axis",fontsize  = 10)
plt.step(x,y,color = "y",marker  = "*")
plt.grid() 
plt.show()
```

```python
# ms parameter : its is basically used to deal with the size of marker parameter.see


x = [1,2,3,4,5]

y = [4,3,6,7,3]
plt.title("Step plot", color = "m") 
plt.xlabel(" X axis",)
plt.ylabel(" Y axis",fontsize  = 10)
plt.step(x,y,color = "y",marker  = "o", ms = 10) # got it!
plt.grid() 
plt.show()
```

```python
# mfc parameter : Its is used to deal with the color of marker.




x = [1,2,3,4,5]

y = [4,3,6,7,3]
plt.title("Step plot", color = "m") 
plt.xlabel(" X axis",)
plt.ylabel(" Y axis",fontsize  = 10)
plt.step(x,y,color = "y",marker  = "o", ms = 10,mfc = "r") # got it!
plt.grid() 
plt.show()
```

```python
# lable ( within the step plot interface) : its works with the plt.legend()




x = [1,2,3,4,5]

y = [4,3,6,7,3]
plt.title("Step plot", color = "m") 
plt.xlabel(" X axis",)
plt.ylabel(" Y axis",fontsize  = 10)
plt.step(x,y,color = "y",marker  = "o", ms = 10,mfc = "r", label = "Python") # got it!
plt.legend()
plt.grid() 
plt.show()
```

```python
# loc = n ( n is integer); this parameter passes in plt.legend() and remember it is uses to change poritions of interrior label.




x = [1,2,3,4,5]

y = [4,3,6,7,3]
plt.title("Step plot", color = "m") 
plt.xlabel(" X axis",)
plt.ylabel(" Y axis",fontsize  = 10)
plt.step(x,y,color = "y",marker  = "o", ms = 10,mfc = "r", label = "Python") # got it!
plt.legend(loc = 2)
plt.show()
```
